```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
convert(df.iloc[0]['genres'])
```

```python
df['genres'] = df['genres'].apply(convert)
df
```

```python
df.loc[0].keywords
```

```python
df['keywords'] = df['keywords'].apply(convert)
df
```

```python
df.iloc[0].cast
```

```python
def convert10(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 10:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
df['cast'] = df['cast'].apply(convert10)
df
```

```python
df.iloc[3]['cast']
```

```python
df.iloc[0]['crew']
```

```python
def find_jobs(obj):
    L = []
    for i in ast.literal_eval(obj):
        if(i['job'] == 'Director'):
            L.append(i['name'])
    return L
```

```python
find_jobs(df.loc[0].crew)
```

```python
df['crew'] = df['crew'].apply(find_jobs)
df
```

### Performing Stemming

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

### Vectorizing the tags of every movie while eliminating the english stop words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 10000 , stop_words = 'english')
```

### Store the vectorized tags of each movie in the variable vectors

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

### Calculating Cosine Similarity between each pair of vectors

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity
```

```python
similarity.shape
```

```python
similarity[0]
```

### Recommending movies based on Cosine Similarity

#### Fetching the index of a movie given its title

```python
movie_index = new_df[new_df['title'] == 'Avatar'].index[0]
movie_index
```

#### Finding the similarity score of the given movie with all the other movies

```python
similarity[movie_index]
```

```python
type(similarity[movie_index])
```

#### Enumerating the array obtained above

```python
enumerate(similarity[movie_index])
```

#### Obtaining a list of pairs of index and value

```python
list(enumerate(similarity[movie_index]))
```

#### Sorting the above list in decreasing order based on the similarity scores

```python
most_similar_movies = sorted(list(enumerate(similarity[movie_index])) , reverse = True , key = lambda x : x[1])
most_similar_movies
```

#### Fetching the indices of the top-5 movies except the given movie itself from the sorted list 

```python
top5_most_similar_movies_indices = []
for i in range(1 , 6):
    top5_most_similar_movies_indices.append(most_similar_movies[i][0])
top5_most_similar_movies_indices
```

#### Fetching the titles of the top-5 most similar movies

```python
top5_most_similar_movies_titles = []
for i in top5_most_similar_movies_indices:
    top5_most_similar_movies_titles.append(new_df['title'][i])
top5_most_similar_movies_titles
```

```python
def recommend(movie):
    
    movie_index = new_df[new_df['title'] == movie].index[0]
    similarity_scores = similarity[movie_index]
    
    most_similar_movies = sorted(list(enumerate(similarity[movie_index])) , reverse = True , key = lambda x : x[1])
    
    top5_most_similar_movies_indices = []
    for i in range(1 , 6):
        top5_most_similar_movies_indices.append(most_similar_movies[i][0])
        
    top5_most_similar_movies_titles = []
    for i in top5_most_similar_movies_indices:
        top5_most_similar_movies_titles.append(new_df['title'][i])
    
    for i in top5_most_similar_movies_titles:
        print(i)
```

```python
recommend('Batman Begins')
```
